## Transaction Frequency vs Average Transaction Amount per Account


```python
# Calculating transaction frequency and average transaction amount per account
account_summary = df.groupby('AccountID').agg({
    'TransactionID': 'count',      # frekuensi transaksi per akun
    'TransactionAmount': 'mean'    # rata-rata jumlah transaksi
}).reset_index()
account_summary.rename(columns={'TransactionID': 'TransactionFrequency'}, inplace=True)
```

```python
# Calculating global average
avg_freq = account_summary['TransactionFrequency'].mean()
avg_amount = account_summary['TransactionAmount'].mean()
```

```python
# Clustering
def quadrant(row):
    if row['TransactionFrequency'] >= avg_freq and row['TransactionAmount'] >= avg_amount:
        return 'Q1: High Freq & High Amount'
    elif row['TransactionFrequency'] < avg_freq and row['TransactionAmount'] >= avg_amount:
        return 'Q2: Low Freq & High Amount'
    elif row['TransactionFrequency'] < avg_freq and row['TransactionAmount'] < avg_amount:
        return 'Q3: Low Freq & Low Amount'
    else:
        return 'Q4: High Freq & Low Amount'

account_summary['Quadrant'] = account_summary.apply(quadrant, axis=1)
```

```python
# Color map
quadrant_colors = {
    'Q1: High Freq & High Amount': '#1A535C',  # merah muda -> potensi fraud tinggi
    'Q2: Low Freq & High Amount': '#FF2E00',   # oranye -> transaksi besar tapi jarang
    'Q3: Low Freq & Low Amount': '#FFA600',    # hijau kebiruan -> akun normal
    'Q4: High Freq & Low Amount': '#00A7E1'    # biru tua -> aktivitas tinggi tapi nilai kecil
}
```

```python
# Setup
fig, ax = plt.subplots(figsize=(16, 6))
fig.suptitle('Transaction Insights by Account', fontsize=18, fontweight='bold')

sns.scatterplot(
    x='TransactionFrequency',
    y='TransactionAmount',
    data=account_summary,
    hue='Quadrant',
    palette=quadrant_colors,
    s=80, alpha=0.75,
    ax=ax
)

ax.text(avg_freq * 1.01, ax.get_ylim()[1]*0.97, f'Avg Freq = {avg_freq:.1f}', color='black')
ax.text(ax.get_xlim()[1]*0.9, avg_amount * 0.9, f'Avg Amount = {avg_amount:.2f}', color='black')

ax.set_title('Transaction Frequency vs Avg Transaction Amount per Account (Quadrant Analysis)',
             fontsize=14, fontweight='bold')
ax.set_xlabel('Transaction Frequency (per Account)')
ax.set_ylabel('Average Transaction Amount')

plt.legend(title='Quadrant', loc='upper left', bbox_to_anchor=(1, 1))

# Cluster line
ax.axvline(avg_freq, color='black', linestyle='--', lw=1.2)
ax.axhline(avg_amount, color='black', linestyle='--', lw=1.2)

# Plot
plt.tight_layout()
plt.show()
```

## Account Distribution By Transaction Frequency and Amount

```python
quadrant_counts = account_summary['Quadrant'].value_counts().reset_index()
quadrant_counts.columns = ['Quadrant', 'Count']
```

```python
# Color map
colors = [
    quadrant_colors[q] for q in quadrant_counts['Quadrant']
]
```

```python
# Pie chart
fig, ax = plt.subplots(figsize=(8, 6))
wedges, texts, autotexts = ax.pie(
    quadrant_counts['Count'],
    labels=quadrant_counts['Quadrant'],
    autopct='%1.1f%%',
    startangle=90,
    colors=colors,
    textprops={'fontsize': 11, 'color': 'black'}
)

# Total account ploting
centre_circle = plt.Circle((0, 0), 0.2, fc='white')  # fc = facecolor
ax.add_artist(centre_circle)

total_accounts = len(account_summary)
ax.text(0, 0, f'Total\n{total_accounts}',
        ha='center', va='center',
        fontsize=11, color='black')

# Plot
plt.setp(autotexts, size=11, weight="bold", color="white")
ax.set_title('Distribution of Accounts by Transaction Quadrant', fontsize=14, fontweight='bold')

plt.show()
```
